# Consolidate into a Single "Is Holiday" Feature

```python
# --- Code for Step 1 ---

# List of your new holiday columns (matching your df_4 columns)
holiday_cols = [
    'is_holiday_Karnataka', 'is_holiday_Tamil_Nadu', 'is_holiday_Andhra_Pradesh',
    'is_holiday_Maharashtra_and_Goa', 'is_holiday_Rest_of_North',
    'is_holiday_Madhya_Pradesh', 'is_holiday_Delhi', 'is_holiday_Rajasthan',
    'is_holiday_East_1', 'is_holiday_Kerala'
]

# Create a master holiday flag using a more efficient pandas method
df_4['is_impacted_by_holiday'] = 0
for col in holiday_cols:
    # This condition flags a row if the holiday matches EITHER the source OR destination region
    is_relevant_holiday = (df_4['srcid_region'] == col.replace('is_holiday_', '')) | \
                          (df_4['destid_region'] == col.replace('is_holiday_', ''))

    # Set the master flag where the holiday is relevant AND flagged
    df_4.loc[is_relevant_holiday & (df_4[col] == 1), 'is_impacted_by_holiday'] = 1


# Let's see how many journeys are affected
print("Journeys impacted by long/important holidays:")
print(df_4['is_impacted_by_holiday'].value_counts())
print("\n" + "="*50 + "\n")
```

```python
# --- Code for Step 2 ---

# 1. Descriptive Statistics
print("Demand statistics grouped by holiday impact:")
print(df_4.groupby('is_impacted_by_holiday')['final_seatcount'].describe())
print("\n" + "="*50 + "\n")

# 2. Visualization (Box Plot)
plt.figure(figsize=(10, 7))
sns.boxplot(data=df_4, x='is_impacted_by_holiday', y='final_seatcount')
plt.title('Final Seat Count for Holiday-Impacted vs. Non-Holiday Journeys', fontsize=16)
plt.xlabel('Is Impacted by Long/Important Holiday', fontsize=12)
plt.ylabel('Final Seat Count', fontsize=12)
plt.xticks([0, 1], ['No', 'Yes'])
plt.grid(True)
plt.show()

# 3. Visualization (KDE Plot)
plt.figure(figsize=(12, 7))
sns.kdeplot(df_4[df_4['is_impacted_by_holiday'] == 0]['final_seatcount'], label='Non-Holiday', fill=True, alpha=0.5)
sns.kdeplot(df_4[df_4['is_impacted_by_holiday'] == 1]['final_seatcount'], label='Holiday-Impacted', fill=True, alpha=0.5)
plt.title('Distribution of Final Seat Count by Holiday Impact', fontsize=16)
plt.xlabel('Final Seat Count', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.show()
```

```python
# --- Function to Add Dynamic Time-Series Features ---

def add_dynamic_features(base_df, transactions_df):
    """
    Enriches the base_df with lag and rolling features from transactions.

    Args:
    - base_df: A dataframe with unique doj-srcid-destid, and a 'dbd' column (set to 15).
    - transactions_df: The full transactions dataframe.

    Returns:
    - An enriched dataframe with new features.
    """
    print("--- Generating dynamic features ---")

    # Sort transactions to ensure correct diff() and rolling() calculations
    transactions_df = transactions_df.sort_values(by=['doj', 'srcid', 'destid', 'dbd'], ascending=[True, True, True, False])

    # 1. Calculate Daily Seats and Searches (the change from the previous day)
    transactions_df['daily_seats'] = transactions_df.groupby(['doj', 'srcid', 'destid'])['cumsum_seatcount'].diff().fillna(0)
    transactions_df['daily_searches'] = transactions_df.groupby(['doj', 'srcid', 'destid'])['cumsum_searchcount'].diff().fillna(0)

    # Clip negative values that might arise from data corrections
    transactions_df['daily_seats'] = transactions_df['daily_seats'].clip(lower=0)
    transactions_df['daily_searches'] = transactions_df['daily_searches'].clip(lower=0)

    enriched_df = base_df.copy()

    # 2. Generate Lag Features
    # We want to know the daily sales for the days right before our prediction point (dbd=15)
    print("Creating lag features for dbd=16, 17, 18, and 22 (one week prior)...")
    for lag in [1, 2, 3, 7]:
        # Prepare a temporary dataframe from transactions
        temp_df = transactions_df[['doj', 'srcid', 'destid', 'dbd', 'daily_seats', 'daily_searches']].copy()

        # Shift the 'dbd' so we can merge it onto our dbd=15 data
        # e.g., for lag=1, we are getting data from dbd=16 and merging it onto our dbd=15 rows.
        temp_df['dbd'] = temp_df['dbd'] - lag

        # Rename columns to be specific
        temp_df.rename(columns={
            'daily_seats': f'daily_seats_lag_{lag}',
            'daily_searches': f'daily_searches_lag_{lag}'
        }, inplace=True)

        # Merge this lag data into our main dataframe
        enriched_df = pd.merge(enriched_df, temp_df, on=['doj', 'srcid', 'destid', 'dbd'], how='left')

    # 3. Generate Rolling Window Features
    print("Creating rolling window features (7-day stats before dbd=15)...")

    # We want stats for the window of dbd=22 to dbd=16
    rolling_window_df = transactions_df[transactions_df['dbd'].between(16, 22)].copy()

    # Group by journey and calculate rolling stats
    rolling_stats = rolling_window_df.groupby(['doj', 'srcid', 'destid']).agg(
        rolling_avg_seats_7d=('daily_seats', 'mean'),
        rolling_std_seats_7d=('daily_seats', 'std'),
        rolling_avg_searches_7d=('daily_searches', 'mean'),
        rolling_std_searches_7d=('daily_searches', 'std')
    ).reset_index()

    # Merge these aggregated stats back into our main dataframe
    enriched_df = pd.merge(enriched_df, rolling_stats, on=['doj', 'srcid', 'destid'], how='left')

    # Fill NaNs that result from merges (e.g., journeys with no data in those lag/rolling periods)
    enriched_df.fillna(0, inplace=True)

    print("Dynamic features generated successfully.")
    return enriched_df
```

```python
# --- Final Pipeline with Dynamic Features ---

# --- Step 1: Load Data & Prepare Lookups (Same as before) ---
print("Loading all data assets...")
df_4 = pd.read_csv("/content/drive/MyDrive/RedBus/updated_model_df_with_aggregated_holidays.csv")
test_df = pd.read_csv("/content/drive/MyDrive/RedBus/test_8gqdJqH (1).csv")
test_calendar_df = pd.read_csv("/content/drive/MyDrive/RedBus/corrected_holiday_encoded.csv")
train_raw_df = pd.read_csv("/content/drive/MyDrive/RedBus/train.csv")
transactions_df = pd.read_csv("/content/drive/MyDrive/RedBus/transactions.csv")

# Convert date columns
df_4['doj'] = pd.to_datetime(df_4['doj'])
test_df['doj'] = pd.to_datetime(test_df['doj'])
transactions_df['doj'] = pd.to_datetime(transactions_df['doj'])
test_calendar_df.rename(columns={'date': 'doj'}, inplace=True)
test_calendar_df['doj'] = pd.to_datetime(test_calendar_df['doj'], format='%d-%m-%Y')

# Create historical average lookups
route_avg_demand = train_raw_df.groupby(['srcid', 'destid'])['final_seatcount'].mean().to_frame('avg_route_demand').reset_index()
src_avg_demand = train_raw_df.groupby('srcid')['final_seatcount'].mean().to_frame('avg_src_demand').reset_index()

# Standardize Holiday Column Names
holiday_region_names = ['Karnataka', 'Tamil Nadu', 'Andhra Pradesh', 'Maharashtra and Goa', 'Rest of North', 'Madhya Pradesh', 'Delhi', 'Rajasthan', 'East 1', 'Kerala']
rename_dict = {name: f"is_holiday_{name.replace(' ', '_')}" for name in holiday_region_names}
df_4.rename(columns=rename_dict, inplace=True)
test_calendar_df.rename(columns=rename_dict, inplace=True)


# --- Step 2: Create ENRICHED Ground-Truth Training Set ---
print("\n--- Creating Enriched Training Set ---")
# Start with your rich df_4 which already has dbd=15 info, calendar, and historical averages
train_enriched_df = add_dynamic_features(df_4, transactions_df)


# --- Step 3: Train the "Pre-Models" on the ENRICHED Data ---
print("\n--- Training ENHANCED Pre-Models ---")

# Define the new, richer feature set for the pre-models
pre_model_features = [
    'srcid', 'destid', 'doj_day_of_week', 'doj_month', 'doj_week_of_year',
    'doj_is_weekend', 'avg_route_demand', 'avg_src_demand'
] + list(rename_dict.values()) + [ # Static Features
    # New Dynamic Features
    'daily_seats_lag_1', 'daily_searches_lag_1',
    'daily_seats_lag_2', 'daily_searches_lag_2',
    'daily_seats_lag_3', 'daily_searches_lag_3',
    'daily_seats_lag_7', 'daily_searches_lag_7',
    'rolling_avg_seats_7d', 'rolling_std_seats_7d',
    'rolling_avg_searches_7d', 'rolling_std_searches_7d'
]

X_pre_train = train_enriched_df[pre_model_features]
y_seats_pre_train = train_enriched_df['seats_at_dbd15']
y_searches_pre_train = train_enriched_df['searches_at_dbd15']

# Train models
lgbm_seats = lgb.LGBMRegressor(random_state=42)
lgbm_seats.fit(X_pre_train, y_seats_pre_train)
print("Enriched pre-model for seats trained.")

lgbm_searches = lgb.LGBMRegressor(random_state=42)
lgbm_searches.fit(X_pre_train, y_searches_pre_train)
print("Enriched pre-model for searches trained.")


# --- Step 4: Prepare the Test Set and Predict Missing Values ---
print("\n--- Preparing Test Set and Applying Dynamic Features ---")
# First, prepare the test set with calendar and historical features
test_processed = pd.merge(test_df, test_calendar_df, on='doj', how='left')
test_processed = pd.merge(test_processed, route_avg_demand, on=['srcid', 'destid'], how='left')
test_processed = pd.merge(test_processed, src_avg_demand, on='srcid', how='left')
test_processed.fillna(0, inplace=True) # Fill all NaNs from merges

# Now, add the dynamic features. We need a 'dbd' column for the function to work.
test_processed['dbd'] = 15
test_processed = add_dynamic_features(test_processed, transactions_df)

# Predict the missing dbd=15 metrics using the enhanced pre-models
X_test_pre_model = test_processed[pre_model_features]
test_processed['seats_at_dbd15'] = lgbm_seats.predict(X_test_pre_model)
test_processed['searches_at_dbd15'] = lgbm_searches.predict(X_test_pre_model)

test_processed['seats_at_dbd15'] = test_processed['seats_at_dbd15'].clip(lower=0)
test_processed['searches_at_dbd15'] = test_processed['searches_at_dbd15'].clip(lower=0)
test_processed['search_to_book_ratio'] = test_processed['seats_at_dbd15'] / (test_processed['searches_at_dbd15'] + 1e-6)


# --- Step 5: Final Model Training using K-Fold Cross-Validation ---
print("\n--- Training Final Model with K-Fold CV ---")

final_model_features = pre_model_features + ['seats_at_dbd15', 'searches_at_dbd15', 'search_to_book_ratio']
X_final_train = train_enriched_df[final_model_features]
y_final_train = train_enriched_df['final_seatcount']
X_final_test = test_processed[final_model_features]

from sklearn.model_selection import KFold

kf = KFold(n_splits=5, shuffle=True, random_state=42)
test_predictions_list = []
final_model_params = {'objective': 'regression_l1', 'random_state': 42, 'n_estimators': 1000}

for fold, (train_idx, val_idx) in enumerate(kf.split(X_final_train)):
    print(f"--- FOLD {fold+1} ---")
    X_train_fold, X_val_fold = X_final_train.iloc[train_idx], X_final_train.iloc[val_idx]
    y_train_fold, y_val_fold = y_final_train.iloc[train_idx], y_final_train.iloc[val_idx]

    model = lgb.LGBMRegressor(**final_model_params)
    model.fit(X_train_fold, y_train_fold,
              eval_set=[(X_val_fold, y_val_fold)],
              eval_metric='rmse',
              callbacks=[lgb.early_stopping(50, verbose=False)])

    fold_preds = model.predict(X_final_test)
    test_predictions_list.append(fold_preds)

# Average predictions from all folds
final_predictions = np.mean(test_predictions_list, axis=0)
final_predictions = np.round(final_predictions.clip(min=0)).astype(int)

# Create submission file
submission_df = pd.DataFrame({'route_key': test_df['route_key'], 'final_seatcount': final_predictions})
submission_df.to_csv('submission_dynamic_features_kfold.csv', index=False)

print("\nSubmission file 'submission_dynamic_features_kfold.csv' created successfully!")
print(submission_df.head())
```

```python
# --- Full Pipeline with Optuna & GPU Acceleration ---

import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.model_selection import KFold, train_test_split
from sklearn.metrics import mean_squared_error
import optuna

optuna.logging.set_verbosity(optuna.logging.WARNING)

# --- Load Data & Prepare Lookups (This part is unchanged) ---
# ... [PASTE THE DATA LOADING AND LOOKUP PREPARATION CODE FROM BEFORE] ...
print("Loading all data assets...")
df_4 = pd.read_csv("/content/drive/MyDrive/RedBus/updated_model_df_with_aggregated_holidays.csv")
test_df = pd.read_csv("/content/drive/MyDrive/RedBus/test_8gqdJqH (1).csv")
test_calendar_df = pd.read_csv("/content/drive/MyDrive/RedBus/corrected_holiday_encoded.csv")
train_raw_df = pd.read_csv("/content/drive/MyDrive/RedBus/train.csv")
transactions_df = pd.read_csv("/content/drive/MyDrive/RedBus/transactions.csv")

# Convert date columns
df_4['doj'] = pd.to_datetime(df_4['doj'])
test_df['doj'] = pd.to_datetime(test_df['doj'])
transactions_df['doj'] = pd.to_datetime(transactions_df['doj'])
test_calendar_df.rename(columns={'date': 'doj'}, inplace=True)
test_calendar_df['doj'] = pd.to_datetime(test_calendar_df['doj'], format='%d-%m-%Y')

# Prepare lookups
route_avg_demand = train_raw_df.groupby(['srcid', 'destid'])['final_seatcount'].mean().to_frame('avg_route_demand').reset_index()
src_avg_demand = train_raw_df.groupby('srcid')['final_seatcount'].mean().to_frame('avg_src_demand').reset_index()

# Rename holiday columns consistently
holiday_region_names = ['Karnataka', 'Tamil Nadu', 'Andhra Pradesh', 'Maharashtra and Goa', 'Rest of North', 'Madhya Pradesh', 'Delhi', 'Rajasthan', 'East 1', 'Kerala']
rename_dict = {name: f"is_holiday_{name.replace(' ', '_')}" for name in holiday_region_names}
df_4.rename(columns=rename_dict, inplace=True)
test_calendar_df.rename(columns=rename_dict, inplace=True)


# --- Define Dynamic Feature Function (Unchanged) ---
def add_dynamic_features(base_df, transactions_df):
    """Enriches the base_df with lag and rolling features from transactions."""
    print("--- Generating dynamic features ---")
    transactions_df = transactions_df.sort_values(by=['doj', 'srcid', 'destid', 'dbd'], ascending=[True, True, True, False])
    transactions_df['daily_seats'] = transactions_df.groupby(['doj', 'srcid', 'destid'])['cumsum_seatcount'].diff().fillna(0).clip(lower=0)
    transactions_df['daily_searches'] = transactions_df.groupby(['doj', 'srcid', 'destid'])['cumsum_searchcount'].diff().fillna(0).clip(lower=0)
    enriched_df = base_df.copy()
    for lag in [1, 2, 3, 7]:
        temp_df = transactions_df[['doj', 'srcid', 'destid', 'dbd', 'daily_seats', 'daily_searches']].copy()
        temp_df['dbd'] -= lag
        temp_df.rename(columns={'daily_seats': f'daily_seats_lag_{lag}', 'daily_searches': f'daily_searches_lag_{lag}'}, inplace=True)
        enriched_df = pd.merge(enriched_df, temp_df, on=['doj', 'srcid', 'destid', 'dbd'], how='left')
    rolling_window_df = transactions_df[transactions_df['dbd'].between(16, 22)].copy()
    rolling_stats = rolling_window_df.groupby(['doj', 'srcid', 'destid']).agg(
        rolling_avg_seats_7d=('daily_seats', 'mean'), rolling_std_seats_7d=('daily_seats', 'std'),
        rolling_avg_searches_7d=('daily_searches', 'mean'), rolling_std_searches_7d=('daily_searches', 'std')
    ).reset_index()
    enriched_df = pd.merge(enriched_df, rolling_stats, on=['doj', 'srcid', 'destid'], how='left')
    enriched_df.fillna(0, inplace=True)
    return enriched_df

# Create the enriched training set
train_enriched_df = add_dynamic_features(df_4, transactions_df)


# --- Step 3: Hyperparameter Tuning with Optuna (GPU Accelerated) ---

# Prepare final training data
final_model_features = pre_model_features = [
    'srcid', 'destid', 'doj_day_of_week', 'doj_month', 'doj_week_of_year',
    'doj_is_weekend', 'avg_route_demand', 'avg_src_demand'
] + list(rename_dict.values()) + [
    'daily_seats_lag_1', 'daily_searches_lag_1', 'daily_seats_lag_2', 'daily_searches_lag_2',
    'daily_seats_lag_3', 'daily_searches_lag_3', 'daily_seats_lag_7', 'daily_searches_lag_7',
    'rolling_avg_seats_7d', 'rolling_std_seats_7d', 'rolling_avg_searches_7d', 'rolling_std_searches_7d'
]
X_final_train = train_enriched_df[final_model_features]
y_final_train = train_enriched_df['final_seatcount']

def objective_gpu(trial):
    """Optuna objective function configured for GPU training."""
    param = {
        'objective': 'regression_l1', 'metric': 'rmse',
        'device': 'gpu',  # <<< KEY CHANGE: Tell LGBM to use the GPU
        'gpu_platform_id': 0, # Usually 0
        'gpu_device_id': 0,   # Usually 0
        'n_estimators': trial.suggest_int('n_estimators', 1000, 5000), # Can increase this with GPU
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.05),
        'num_leaves': trial.suggest_int('num_leaves', 31, 128), # Can afford more leaves
        'max_depth': trial.suggest_int('max_depth', 7, 15),
        'min_child_samples': trial.suggest_int('min_child_samples', 20, 100),
        'feature_fraction': trial.suggest_float('feature_fraction', 0.7, 1.0),
        'bagging_fraction': trial.suggest_float('bagging_fraction', 0.7, 1.0),
        'bagging_freq': trial.suggest_int('bagging_freq', 1, 7),
        'lambda_l1': trial.suggest_float('lambda_l1', 1e-8, 1.0, log=True),
        'lambda_l2': trial.suggest_float('lambda_l2', 1e-8, 1.0, log=True),
        'verbose': -1, 'seed': 42
    }

    X_train_tune, X_val_tune, y_train_tune, y_val_tune = train_test_split(X_final_train, y_final_train, test_size=0.2, random_state=42)
    model = lgb.LGBMRegressor(**param)
    model.fit(X_train_tune, y_train_tune,
              eval_set=[(X_val_tune, y_val_tune)],
              eval_metric='rmse',
              callbacks=[lgb.early_stopping(100, verbose=False)]) # Increased patience

    preds = model.predict(X_val_tune)
    rmse = np.sqrt(mean_squared_error(y_val_tune, preds))
    return rmse

# Run the study
print("\n--- Starting GPU-Accelerated Hyperparameter Tuning ---")
study = optuna.create_study(direction='minimize')
study.optimize(objective_gpu, n_trials=100) # Should be much faster now

print("\n--- Tuning Complete ---")
print("Best trial RMSE: ", study.best_value)
print("Best parameters found: ", study.best_params)
best_params = study.best_params


# --- Step 4: Final Model Training using BEST GPU Parameters and K-Fold ---

# 4.1: Prepare the test set (same logic as before, but ensure it's fully run)
# ... [Paste the full test_processed creation and pre-model prediction code here] ...
# For brevity, assuming test_processed and X_final_test are fully created as in the last step

# 4.2: Train the final model with the BEST params and K-Fold on GPU
print("\n--- Training Final GPU-Accelerated Model with Tuned Parameters and K-Fold CV ---")
kf = KFold(n_splits=5, shuffle=True, random_state=42)
test_predictions_list = []

final_model_params = best_params
final_model_params['objective'] = 'regression_l1'
final_model_params['random_state'] = 42
final_model_params['device'] = 'gpu' # <<< KEY CHANGE: Use GPU for final training too

for fold, (train_idx, val_idx) in enumerate(kf.split(X_final_train)):
    print(f"--- FOLD {fold+1} ---")
    X_train_fold, y_train_fold = X_final_train.iloc[train_idx], y_final_train.iloc[train_idx]

    model = lgb.LGBMRegressor(**final_model_params)
    model.fit(X_train_fold, y_train_fold) # No need for eval set here, we use the full fold

    fold_preds = model.predict(X_final_test)
    test_predictions_list.append(fold_preds)

# 4.3: Generate Submission
final_predictions = np.mean(test_predictions_list, axis=0)
final_predictions = np.round(final_predictions.clip(lower=0)).astype(int)

submission_df = pd.DataFrame({'route_key': test_df['route_key'], 'final_seatcount': final_predictions})
submission_df.to_csv('submission_tuned_gpu_kfold.csv', index=False)

print("\nSubmission file 'submission_tuned_gpu_kfold.csv' created successfully!")
print(submission_df.head())
```

```python
# --- FINAL ROBUST SUBMISSION SCRIPT (V5 - Consistent Feature Naming) ---

import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.model_selection import KFold
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

def run_full_pipeline():
    """
    This function encapsulates the entire workflow with corrected variable names and logic.
    Ensures consistent feature naming between train and test.
    """
    # --- Step 1: Load All Data Assets ---
    print("--- Step 1: Loading all data assets... ---")
    try:
        # Renamed for clarity - this is the pre-processed train data with dbd15 stats and regions
        train_base_df = pd.read_csv("/content/drive/MyDrive/RedBus/updated_model_df_with_aggregated_holidays.csv")
        test_df_base = pd.read_csv("/content/drive/MyDrive/RedBus/test_8gqdJqH (1).csv")
        calendar_features_df = pd.read_csv("/content/drive/MyDrive/RedBus/corrected_holiday_encoded.csv") # Using one name for clarity
        train_raw_df = pd.read_csv("/content/drive/MyDrive/RedBus/train.csv") # For historical averages
        transactions_df = pd.read_csv("/content/drive/MyDrive/RedBus/transactions.csv") # For dynamic features & region lookup
        print("All data files loaded successfully.")
    except FileNotFoundError as e:
        print(f"ERROR: Could not find a data file. Please check path. Details: {e}")
        return

    # --- Step 2: Prepare Common Data Assets & Lookups ---
    print("\n--- Step 2: Preparing common data assets... ---")

    # Convert date columns (doj) for all relevant dataframes
    for df in [train_base_df, test_df_base, transactions_df]:
        if 'doj' in df.columns:
            df['doj'] = pd.to_datetime(df['doj'])

    calendar_features_df.rename(columns={'date': 'doj'}, inplace=True, errors='ignore')
    calendar_features_df['doj'] = pd.to_datetime(calendar_features_df['doj'], format='%d-%m-%Y', errors='coerce')

    # Create Lookups from historical data (train_raw_df and transactions_df)
    route_avg_demand = train_raw_df.groupby(['srcid', 'destid'])['final_seatcount'].mean().to_frame('avg_route_demand').reset_index()
    src_avg_demand = train_raw_df.groupby('srcid')['final_seatcount'].mean().to_frame('avg_src_demand').reset_index()

    # Ensure region_tier_lookup comes from transactions_df as it contains this info
    region_tier_lookup = transactions_df[['srcid', 'destid', 'srcid_region', 'destid_region', 'srcid_tier', 'destid_tier']].drop_duplicates(subset=['srcid', 'destid'])

    # Standardize Holiday Column Names in the calendar data (it will be merged later)
    holiday_region_names = ['Karnataka', 'Tamil Nadu', 'Andhra Pradesh', 'Maharashtra and Goa', 'Rest of North', 'Madhya Pradesh', 'Delhi', 'Rajasthan', 'East 1', 'Kerala']
    rename_dict = {name: f"is_holiday_{name.replace(' ', '_')}" for name in holiday_region_names}

    # Apply rename to calendar_features_df
    # First, filter for columns that actually need renaming based on the dictionary keys
    cols_to_rename = {col: rename_dict[col.replace('_is_holiday', '')] for col in calendar_features_df.columns if '_is_holiday' in col and col.replace('_is_holiday', '') in holiday_region_names}
    calendar_features_df.rename(columns=cols_to_rename, inplace=True)

    # Ensure calendar_features_df only contains doj and the standardized holiday columns
    calendar_holiday_cols = ['doj'] + list(rename_dict.values())
    calendar_features_df = calendar_features_df[calendar_holiday_cols].copy()


    # --- Step 3: Define and Apply Feature Engineering Steps Consistently ---
    print("\n--- Step 3: Processing train and test dataframes... ---")

    # Start with the base dataframes
    train_processed = train_base_df.copy()
    test_processed = test_df_base.copy()

    # Merge calendar/holiday features (now with standardized names)
    # No suffixes needed now as calendar columns are already renamed
    train_processed = pd.merge(train_processed, calendar_features_df, on='doj', how='left')
    test_processed = pd.merge(test_processed, calendar_features_df, on='doj', how='left')

    # Merge region/tier for test set (train_processed already has it)
    test_processed = pd.merge(test_processed, region_tier_lookup, on=['srcid', 'destid'], how='left')

    # Merge historical averages
    train_processed = pd.merge(train_processed, route_avg_demand, on=['srcid', 'destid'], how='left')
    train_processed = pd.merge(train_processed, src_avg_demand, on='srcid', how='left')
    test_processed = pd.merge(test_processed, route_avg_demand, on=['srcid', 'destid'], how='left')
    test_processed = pd.merge(test_processed, src_avg_demand, on='srcid', how='left')

    # Add dynamic (lag/rolling) features
    # Re-process transactions to ensure correct lag/rolling calculation based on DOJ
    temp_transactions = transactions_df.copy()
    temp_transactions = temp_transactions.sort_values(by=['doj', 'srcid', 'destid', 'dbd'], ascending=[True, True, True, False])
    temp_transactions['daily_seats'] = temp_transactions.groupby(['doj', 'srcid', 'destid'])['cumsum_seatcount'].diff().fillna(0).clip(lower=0)
    temp_transactions['daily_searches'] = temp_transactions.groupby(['doj', 'srcid', 'destid'])['cumsum_searchcount'].diff().fillna(0).clip(lower=0)

    # Define the point in time (dbd=15) for the base dataframes
    train_processed['dbd'] = 15
    test_processed['dbd'] = 15

    # Add lag features
    for lag in [1, 2, 3, 7]:
        temp_df = temp_transactions[['doj', 'srcid', 'destid', 'dbd', 'daily_seats', 'daily_searches']].copy()
        temp_df['dbd'] -= lag # This is the dbd value we are joining from
        temp_df.rename(columns={'daily_seats': f'daily_seats_lag_{lag}', 'daily_searches': f'daily_searches_lag_{lag}'}, inplace=True)
        train_processed = pd.merge(train_processed, temp_df, on=['doj', 'srcid', 'destid', 'dbd'], how='left')
        test_processed = pd.merge(test_processed, temp_df, on=['doj', 'srcid', 'destid', 'dbd'], how='left')

    # Add rolling features (stats from dbd 16 to 22, joining onto dbd 15)
    rolling_window_df = temp_transactions[temp_transactions['dbd'].between(16, 22)].copy()
    rolling_stats = rolling_window_df.groupby(['doj', 'srcid', 'destid']).agg(
        rolling_avg_seats_7d=('daily_seats', 'mean'), rolling_std_seats_7d=('daily_seats', 'std'),
        rolling_avg_searches_7d=('daily_searches', 'mean'), rolling_std_searches_7d=('daily_searches', 'std')
    ).reset_index()

    # Since rolling stats are calculated on a date/route basis independent of the join 'dbd',
    # we merge them directly onto the date/route keys.
    train_processed = pd.merge(train_processed, rolling_stats, on=['doj', 'srcid', 'destid'], how='left')
    test_processed = pd.merge(test_processed, rolling_stats, on=['doj', 'srcid', 'destid'], how='left')


    # --- Final Cleanup ---
    print("Performing final cleanup (filling NaNs, encoding)...")
    # Fill NaNs from merges (e.g., new routes, missing lag/rolling data)
    # Use global train mean for avg_demand NaNs
    train_processed['avg_route_demand'].fillna(train_raw_df['final_seatcount'].mean(), inplace=True)
    train_processed['avg_src_demand'].fillna(train_raw_df['final_seatcount'].mean(), inplace=True)
    test_processed['avg_route_demand'].fillna(train_raw_df['final_seatcount'].mean(), inplace=True)
    test_processed['avg_src_demand'].fillna(train_raw_df['final_seatcount'].mean(), inplace=True)

    # Fill NaNs for lag and rolling features with 0 (assuming no activity)
    lag_rolling_cols = [col for col in train_processed.columns if '_lag_' in col or 'rolling_' in col]
    for col in lag_rolling_cols:
         train_processed[col].fillna(0, inplace=True)
         test_processed[col].fillna(0, inplace=True)

    # Fill region/tier NaNs in test set (from new routes/cities)
    for col in ['srcid_region', 'destid_region', 'srcid_tier', 'destid_tier']:
        if test_processed[col].isnull().any():
            # Use mode from the combined train+transactions lookup for safety
            mode_val = region_tier_lookup[col].mode()[0] if not region_tier_lookup[col].empty else 'Unknown'
            test_processed[col].fillna(mode_val, inplace=True)

    # Handle any remaining NaNs (e.g., new holiday types not in calendar_features_df)
    train_processed.fillna(0, inplace=True) # Be cautious with this - inspect NaNs first
    test_processed.fillna(0, inplace=True)   # Be cautious with this - inspect NaNs first


    # --- Step 4: Run the Pre-Model Stage ---
    print("\n--- Step 4: Training and applying pre-models... ---")

    # Define pre-model features based on the final processed dataframes
    # Exclude the targets and features that are only known at dbd=15 or later
    pre_model_features = [col for col in train_processed.columns if 'is_holiday' in col or 'doj_' in col or 'avg_' in col or '_lag_' in col or 'rolling_' in col]
    pre_model_features += ['srcid', 'destid'] # Include srcid and destid as categorical/ID features

    # Ensure no target or post-dbd15 features are included
    pre_model_features = [f for f in pre_model_features if f not in ['final_seatcount', 'seats_at_dbd15', 'searches_at_dbd15', 'search_to_book_ratio', 'dbd']]

    # Ensure both dataframes have all the pre_model_features columns
    missing_in_train = set(pre_model_features) - set(train_processed.columns)
    missing_in_test = set(pre_model_features) - set(test_processed.columns)
    if missing_in_train: print(f"Warning: Missing pre-model features in train_processed: {missing_in_train}")
    if missing_in_test: print(f"Warning: Missing pre-model features in test_processed: {missing_in_test}")

    # Filter to common features only for robustness
    common_pre_model_features = list(set(pre_model_features) & set(train_processed.columns) & set(test_processed.columns))

    X_pre_train = train_processed[common_pre_model_features]
    X_test_pre_model = test_processed[common_pre_model_features]

    y_seats_pre_train = train_processed['seats_at_dbd15']
    y_searches_pre_train = train_processed['searches_at_dbd15']

    lgbm_seats = lgb.LGBMRegressor(random_state=42); lgbm_seats.fit(X_pre_train, y_seats_pre_train)
    lgbm_searches = lgb.LGBMRegressor(random_state=42); lgbm_searches.fit(X_pre_train, y_searches_pre_train)

    test_processed['seats_at_dbd15'] = lgbm_seats.predict(X_test_pre_model).clip(lower=0)
    test_processed['searches_at_dbd15'] = lgbm_searches.predict(X_test_pre_model).clip(lower=0)
    test_processed['search_to_book_ratio'] = test_processed['seats_at_dbd15'] / (test_processed['searches_at_dbd15'] + 1e-6)
    print("Pre-model predictions generated for test set.")

    # --- Step 5: Final Model Training & Submission ---
    print("\n--- Step 5: Training final model and creating submission... ---")

    # The final model uses the original features PLUS the ones we just predicted/created
    final_model_features = common_pre_model_features + ['seats_at_dbd15', 'searches_at_dbd15', 'search_to_book_ratio']

    X_final_train = train_processed[final_model_features]
    y_final_train = train_processed['final_seatcount']
    X_final_test = test_processed[final_model_features]

    # Your tuned parameters for CPU
    best_params_cpu = {
        'objective': 'regression_l1', 'metric': 'rmse', 'random_state': 42, 'n_jobs': -1,
        'n_estimators': 4810, 'learning_rate': 0.02794599333522594, 'num_leaves': 121,
        'max_depth': 15, 'min_child_samples': 30, 'feature_fraction': 0.7166239010671843,
        'bagging_fraction': 0.8765482588654137, 'bagging_freq': 1, 'lambda_l1': 3.461019426015012e-05,
        'lambda_l2': 2.9638572416890396e-06
    }

    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    test_predictions_list = []

    for fold, (train_idx, val_idx) in enumerate(kf.split(X_final_train)):
        print(f"--- Training Final Model: FOLD {fold+1}/5 ---")
        X_train_fold, y_train_fold = X_final_train.iloc[train_idx], y_final_train.iloc[train_idx]

        model = lgb.LGBMRegressor(**best_params_cpu)
        model.fit(X_train_fold, y_train_fold)

        fold_preds = model.predict(X_final_test)
        test_predictions_list.append(fold_preds)

    final_predictions = np.mean(test_predictions_list, axis=0)
    final_predictions = np.round(final_predictions.clip(lower=0)).astype(int)

    submission_df = pd.DataFrame({'route_key': test_df_base['route_key'], 'final_seatcount': final_predictions})
    submission_df.to_csv('submission_final_robust_cpu_v5.csv', index=False)

    print("\n\n--- SCRIPT COMPLETE ---")
    print("Submission file 'submission_final_robust_cpu_v5.csv' created successfully!")
    print("\nSample of your submission:")
    print(submission_df.head())


# --- Run the entire process ---
run_full_pipeline()
```

```python
# --- FINAL ENSEMBLE SUBMISSION SCRIPT (LGBM + XGBoost) ---

import pandas as pd
import numpy as np
import lightgbm as lgb
import xgboost as xgb
from sklearn.model_selection import KFold
import warnings

# Suppress common pandas and LightGBM warnings for cleaner output
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=UserWarning)

def run_full_ensemble_pipeline():
    """
    This function encapsulates the entire workflow: data loading, advanced feature engineering,
    a two-stage pre-model process, and a final K-Fold ensemble of LGBM and XGBoost.
    """

    # --- Step 1: Load All Data Assets ---
    print("--- Step 1: Loading all data assets... ---")
    try:
        df_4 = pd.read_csv("/content/drive/MyDrive/RedBus/updated_model_df_with_aggregated_holidays.csv")
        test_df = pd.read_csv("/content/drive/MyDrive/RedBus/test_8gqdJqH (1).csv")
        test_calendar_df = pd.read_csv("/content/drive/MyDrive/RedBus/corrected_holiday_encoded.csv")
        train_raw_df = pd.read_csv("/content/drive/MyDrive/RedBus/train.csv")
        transactions_df = pd.read_csv("/content/drive/MyDrive/RedBus/transactions.csv")
        print("All data files loaded successfully.")
    except FileNotFoundError as e:
        print(f"ERROR: Could not find a data file. Please check path. Details: {e}")
        return

    # --- Step 2: Define the Advanced Feature Engineering Function ---
    def add_dynamic_features(base_df, transactions_df, route_avgs, src_avgs):
        """Enriches a dataframe with lag, rolling, and other advanced features."""
        enriched_df = base_df.copy()

        # Merge historical averages first, so we can use them in ratio features
        enriched_df = pd.merge(enriched_df, route_avgs, on=['srcid', 'destid'], how='left')
        enriched_df = pd.merge(enriched_df, src_avgs, on='srcid', how='left')
        enriched_df['avg_route_demand'].fillna(train_raw_df['final_seatcount'].mean(), inplace=True)
        enriched_df['avg_src_demand'].fillna(train_raw_df['final_seatcount'].mean(), inplace=True)

        # Ensure transactions date column is in the right format
        transactions_df['doj'] = pd.to_datetime(transactions_df['doj'])

        # Calculate daily changes
        temp_transactions = transactions_df.sort_values(by=['doj', 'srcid', 'destid', 'dbd'], ascending=[True, True, True, False])
        temp_transactions['daily_seats'] = temp_transactions.groupby(['doj', 'srcid', 'destid'])['cumsum_seatcount'].diff().fillna(0).clip(lower=0)
        temp_transactions['daily_searches'] = temp_transactions.groupby(['doj', 'srcid', 'destid'])['cumsum_searchcount'].diff().fillna(0).clip(lower=0)

        if 'dbd' not in enriched_df.columns: enriched_df['dbd'] = 15

        for lag in [1, 2, 3, 7]:
            temp_df = temp_transactions[['doj', 'srcid', 'destid', 'dbd', 'daily_seats', 'daily_searches']].copy()
            temp_df['dbd'] -= lag
            temp_df.rename(columns={'daily_seats': f'daily_seats_lag_{lag}', 'daily_searches': f'daily_searches_lag_{lag}'}, inplace=True)
            enriched_df = pd.merge(enriched_df, temp_df, on=['doj', 'srcid', 'destid', 'dbd'], how='left')

        rolling_window_df = temp_transactions[temp_transactions['dbd'].between(16, 22)].copy()
        rolling_stats = rolling_window_df.groupby(['doj', 'srcid', 'destid']).agg(
            rolling_avg_seats_7d=('daily_seats', 'mean'), rolling_std_seats_7d=('daily_seats', 'std'),
            rolling_max_seats_7d=('daily_seats', 'max'), rolling_skew_seats_7d=('daily_seats', 'skew')
        ).reset_index()
        enriched_df = pd.merge(enriched_df, rolling_stats, on=['doj', 'srcid', 'destid'], how='left')

        enriched_df.fillna(0, inplace=True)

        enriched_df['daily_seats_velocity_1d'] = enriched_df['daily_seats_lag_1'] - enriched_df['daily_seats_lag_2']
        enriched_df['daily_seats_acceleration_1d'] = enriched_df['daily_seats_velocity_1d'] - (enriched_df['daily_seats_lag_2'] - enriched_df['daily_seats_lag_3'])
        enriched_df['lag1_vs_route_avg_ratio'] = enriched_df['daily_seats_lag_1'] / ((enriched_df['avg_route_demand'] / 30) + 1e-6)

        return enriched_df

    # --- Step 3: Prepare Base Training and Test Sets ---
    print("\n--- Step 3: Preparing base dataframes... ---")

    # Date conversions and column renaming
    df_4['doj'] = pd.to_datetime(df_4['doj'])
    test_df['doj'] = pd.to_datetime(test_df['doj'])
    test_calendar_df.rename(columns={'date': 'doj'}, inplace=True, errors='ignore')
    test_calendar_df['doj'] = pd.to_datetime(test_calendar_df['doj'], format='%d-%m-%Y', errors='coerce')
    holiday_region_names = ['Karnataka', 'Tamil Nadu', 'Andhra Pradesh', 'Maharashtra and Goa', 'Rest of North', 'Madhya Pradesh', 'Delhi', 'Rajasthan', 'East 1', 'Kerala']
    rename_dict = {name: f"is_holiday_{name.replace(' ', '_')}" for name in holiday_region_names}
    df_4.rename(columns=rename_dict, inplace=True, errors='ignore')
    test_calendar_df.rename(columns=rename_dict, inplace=True, errors='ignore')

    # Create enriched training set
    train_enriched_df = add_dynamic_features(df_4, transactions_df, route_avg_demand, src_avg_demand)

    # Create enriched test set
    test_with_calendar = pd.merge(test_df, test_calendar_df, on='doj', how='left')
    test_enriched_df = add_dynamic_features(test_with_calendar, transactions_df, route_avg_demand, src_avg_demand)

    # --- Step 4: Pre-Model Stage ---
    print("\n--- Step 4: Training and applying pre-models... ---")
    pre_model_features = [col for col in train_enriched_df.columns if
                          'is_holiday' in col or 'doj_' in col or 'avg_' in col or
                          '_lag_' in col or 'rolling_' in col or 'velocity' in col or
                          'acceleration' in col or 'ratio' in col]
    pre_model_features += ['srcid', 'destid']
    pre_model_features = list(set([f for f in pre_model_features if f in train_enriched_df.columns and f not in ['final_seatcount', 'seats_at_dbd15', 'searches_at_dbd15', 'search_to_book_ratio']]))

    X_pre_train = train_enriched_df[pre_model_features]
    X_test_pre_model = test_enriched_df[pre_model_features]

    lgbm_seats = lgb.LGBMRegressor(random_state=42); lgbm_seats.fit(X_pre_train, train_enriched_df['seats_at_dbd15'])
    lgbm_searches = lgb.LGBMRegressor(random_state=42); lgbm_searches.fit(X_pre_train, train_enriched_df['searches_at_dbd15'])

    test_enriched_df['seats_at_dbd15'] = lgbm_seats.predict(X_test_pre_model).clip(lower=0)
    test_enriched_df['searches_at_dbd15'] = lgbm_searches.predict(X_test_pre_model).clip(lower=0)
    test_enriched_df['search_to_book_ratio'] = test_enriched_df['seats_at_dbd15'] / (test_enriched_df['searches_at_dbd15'] + 1e-6)

    # --- Step 5: Final Ensemble Model Training & Submission ---
    print("\n--- Step 5: Training ENSEMBLE model and creating submission... ---")
    final_model_features = pre_model_features + ['seats_at_dbd15', 'searches_at_dbd15', 'search_to_book_ratio']
    X_final_train = train_enriched_df[final_model_features]
    y_final_train = train_enriched_df['final_seatcount']
    X_final_test = test_enriched_df[final_model_features]

    # Your tuned LGBM parameters
    lgbm_params = {
        'objective': 'regression_l1', 'metric': 'rmse', 'random_state': 42, 'n_jobs': -1,
        'n_estimators': 4810, 'learning_rate': 0.02794599333522594, 'num_leaves': 121,
        'max_depth': 15, 'min_child_samples': 30, 'feature_fraction': 0.7166239010671843,
        'bagging_fraction': 0.8765482588654137, 'bagging_freq': 1, 'lambda_l1': 3.461019426015012e-05,
        'lambda_l2': 2.9638572416890396e-06
    }

    # Solid default parameters for XGBoost
    xgb_params = {
        'objective': 'reg:squarederror', 'eval_metric': 'rmse', 'random_state': 42,
        'n_estimators': 2000, 'learning_rate': 0.02, 'max_depth': 8,
        'subsample': 0.8, 'colsample_bytree': 0.8, 'n_jobs': -1
    }

    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    lgbm_test_preds = []
    xgb_test_preds = []

    for fold, (train_idx, val_idx) in enumerate(kf.split(X_final_train)):
        print(f"--- Training ENSEMBLE: FOLD {fold+1}/5 ---")
        X_train_fold, y_train_fold = X_final_train.iloc[train_idx], y_final_train.iloc[train_idx]
        X_val_fold, y_val_fold = X_final_train.iloc[val_idx], y_final_train.iloc[val_idx]

        print("  Training LGBM...")
        lgbm_model = lgb.LGBMRegressor(**lgbm_params)
        lgbm_model.fit(X_train_fold, y_train_fold)
        lgbm_test_preds.append(lgbm_model.predict(X_final_test))

        print("  Training XGBoost...")
        xgb_model = xgb.XGBRegressor(**xgb_params)
        xgb_model.fit(X_train_fold, y_train_fold,
                      eval_set=[(X_val_fold, y_val_fold)],
                      early_stopping_rounds=50,
                      verbose=False)
        xgb_test_preds.append(xgb_model.predict(X_final_test))

    # --- Blend the Predictions ---
    print("\n--- Blending Model Predictions ---")
    lgbm_final_preds = np.mean(lgbm_test_preds, axis=0)
    xgb_final_preds = np.mean(xgb_test_preds, axis=0)
    ensembled_predictions = 0.5 * lgbm_final_preds + 0.5 * xgb_final_preds

    final_predictions = np.round(ensembled_predictions.clip(min=0)).astype(int)

    # --- Create Submission File ---
    submission_df = pd.DataFrame({'route_key': test_df['route_key'], 'final_seatcount': final_predictions})
    submission_df.to_csv('submission_ensemble_final.csv', index=False)

    print("\n\n--- SCRIPT COMPLETE ---")
    print("Submission file 'submission_ensemble_final.csv' created successfully!")
    print("\nSample of your submission:")
    print(submission_df.head())


# --- Run the entire process ---
run_full_ensemble_pipeline()
```

```python
# --- FINAL ROBUST SUBMISSION SCRIPT (Combined Best Practices) ---

import pandas as pd
import numpy as np
import lightgbm as lgb
import xgboost as xgb
from sklearn.model_selection import KFold
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=UserWarning)

# --- This is the function from your first script, now re-integrated ---
def add_dynamic_features(base_df, transactions_df):
    """Enriches the base_df with lag, rolling, and ADVANCED features."""
    print("--- Generating dynamic (lag, rolling, velocity) features ---")

    # Sort transactions to ensure correct diff() and rolling() calculations
    temp_transactions = transactions_df.sort_values(by=['doj', 'srcid', 'destid', 'dbd'], ascending=[True, True, True, False])

    # Calculate Daily Seats and Searches
    temp_transactions['daily_seats'] = temp_transactions.groupby(['doj', 'srcid', 'destid'])['cumsum_seatcount'].diff().fillna(0).clip(lower=0)
    temp_transactions['daily_searches'] = temp_transactions.groupby(['doj', 'srcid', 'destid'])['cumsum_searchcount'].diff().fillna(0).clip(lower=0)

    enriched_df = base_df.copy()

    # Generate Lag Features
    for lag in [1, 2, 3, 7]:
        temp_df = temp_transactions[['doj', 'srcid', 'destid', 'dbd', 'daily_seats', 'daily_searches']].copy()
        temp_df['dbd'] -= lag
        temp_df.rename(columns={'daily_seats': f'daily_seats_lag_{lag}', 'daily_searches': f'daily_searches_lag_{lag}'}, inplace=True)
        enriched_df = pd.merge(enriched_df, temp_df, on=['doj', 'srcid', 'destid', 'dbd'], how='left')

    # Generate Rolling Window Features
    rolling_window_df = temp_transactions[temp_transactions['dbd'].between(16, 22)].copy()
    rolling_stats = rolling_window_df.groupby(['doj', 'srcid', 'destid']).agg(
        rolling_avg_seats_7d=('daily_seats', 'mean'),
        rolling_std_seats_7d=('daily_seats', 'std'),
        rolling_max_seats_7d=('daily_seats', 'max'),
        rolling_skew_seats_7d=('daily_seats', 'skew'),
        rolling_avg_searches_7d=('daily_searches', 'mean'),
        rolling_std_searches_7d=('daily_searches', 'std')
    ).reset_index()
    enriched_df = pd.merge(enriched_df, rolling_stats, on=['doj', 'srcid', 'destid'], how='left')

    # Fill NaNs from lags/rolling before creating velocity features
    enriched_df.fillna(0, inplace=True)

    # Generate Advanced Velocity/Acceleration/Ratio Features
    enriched_df['daily_seats_velocity_1d'] = enriched_df['daily_seats_lag_1'] - enriched_df['daily_seats_lag_2']
    enriched_df['daily_seats_acceleration_1d'] = enriched_df['daily_seats_velocity_1d'] - (enriched_df['daily_seats_lag_2'] - enriched_df['daily_seats_lag_3'])
    enriched_df['lag1_vs_route_avg_ratio'] = enriched_df['daily_seats_lag_1'] / ((enriched_df['avg_route_demand'] / 30) + 1e-6)

    return enriched_df

# --- Step 1: Load All Data Assets ---
print("--- Step 1: Loading all data assets... ---")
df_4 = pd.read_csv("/content/drive/MyDrive/RedBus/updated_model_df_with_aggregated_holidays.csv")
test_df = pd.read_csv("/content/drive/MyDrive/RedBus/test_8gqdJqH (1).csv")
test_calendar_df = pd.read_csv("/content/drive/MyDrive/RedBus/corrected_holiday_encoded.csv")
train_raw_df = pd.read_csv("/content/drive/MyDrive/RedBus/train.csv")
transactions_df = pd.read_csv("/content/drive/MyDrive/RedBus/transactions.csv")
print("All data files loaded successfully.")

# --- Step 2: Prepare Common Data Assets & Lookups ---
print("\n--- Step 2: Preparing common data assets... ---")
df_4['doj'] = pd.to_datetime(df_4['doj'])
test_df['doj'] = pd.to_datetime(test_df['doj'])
transactions_df['doj'] = pd.to_datetime(transactions_df['doj'])
test_calendar_df.rename(columns={'date': 'doj'}, inplace=True, errors='ignore')
test_calendar_df['doj'] = pd.to_datetime(test_calendar_df['doj'], format='%d-%m-%Y', errors='coerce')

route_avg_demand = train_raw_df.groupby(['srcid', 'destid'])['final_seatcount'].mean().to_frame('avg_route_demand').reset_index()
src_avg_demand = train_raw_df.groupby('srcid')['final_seatcount'].mean().to_frame('avg_src_demand').reset_index()

holiday_region_names = ['Karnataka', 'Tamil Nadu', 'Andhra Pradesh', 'Maharashtra and Goa', 'Rest of North', 'Madhya Pradesh', 'Delhi', 'Rajasthan', 'East 1', 'Kerala']
rename_dict = {name: f"is_holiday_{name.replace(' ', '_')}" for name in holiday_region_names}
df_4.rename(columns=rename_dict, inplace=True, errors='ignore')
test_calendar_df.rename(columns=rename_dict, inplace=True, errors='ignore')

# --- Step 3: Engineer Rich Feature Sets for Train and Test ---
print("\n--- Step 3: Engineering rich feature sets... ---")
# Prepare base training data
train_processed = df_4.copy()

# Prepare base test data
test_processed = pd.merge(test_df, test_calendar_df, on='doj', how='left')
test_processed = pd.merge(test_processed, route_avg_demand, on=['srcid', 'destid'], how='left')
test_processed = pd.merge(test_processed, src_avg_demand, on='srcid', how='left')
test_processed['avg_route_demand'].fillna(train_raw_df['final_seatcount'].mean(), inplace=True)
test_processed['avg_src_demand'].fillna(train_raw_df['final_seatcount'].mean(), inplace=True)

# Set dbd=15 for both to generate dynamic features correctly
train_processed['dbd'] = 15
test_processed['dbd'] = 15

# Apply dynamic feature generation to BOTH dataframes
train_processed = add_dynamic_features(train_processed, transactions_df)
test_processed = add_dynamic_features(test_processed, transactions_df)
train_processed.fillna(0, inplace=True)
test_processed.fillna(0, inplace=True)

# --- Step 4: Pre-Model Stage (Leakage-Free Feature Generation) ---
print("\n--- Step 4: Generating leakage-free meta-features with pre-models... ---")
pre_model_features = [
    'srcid', 'destid', 'doj_day_of_week', 'doj_month', 'doj_week_of_year',
    'doj_is_weekend', 'avg_route_demand', 'avg_src_demand'
] + list(rename_dict.values()) + [
    # All dynamic features are now included here for the pre-models
    'daily_seats_lag_1', 'daily_searches_lag_1', 'daily_seats_lag_2', 'daily_searches_lag_2',
    'daily_seats_lag_3', 'daily_searches_lag_3', 'daily_seats_lag_7', 'daily_searches_lag_7',
    'rolling_avg_seats_7d', 'rolling_std_seats_7d', 'rolling_max_seats_7d', 'rolling_skew_seats_7d',
    'rolling_avg_searches_7d', 'rolling_std_searches_7d',
    'daily_seats_velocity_1d', 'daily_seats_acceleration_1d', 'lag1_vs_route_avg_ratio'
]
X_pre_train = train_processed[pre_model_features]
X_test_pre_model = test_processed[pre_model_features]
y_seats_pre_train = train_processed['seats_at_dbd15']
y_searches_pre_train = train_processed['searches_at_dbd15']

# Use Out-of-Fold prediction to generate features for the training set (to prevent leakage)
print("  Generating OOF predictions for training set...")
train_processed['seats_at_dbd15_pred'] = np.nan
train_processed['searches_at_dbd15_pred'] = np.nan
kf_pre_model = KFold(n_splits=5, shuffle=True, random_state=42)
for train_idx, val_idx in kf_pre_model.split(X_pre_train):
    X_train_fold, X_val_fold = X_pre_train.iloc[train_idx], X_pre_train.iloc[val_idx]

    lgbm_s = lgb.LGBMRegressor(random_state=42).fit(X_train_fold, y_seats_pre_train.iloc[train_idx])
    train_processed.loc[train_processed.index[val_idx], 'seats_at_dbd15_pred'] = lgbm_s.predict(X_val_fold)

    lgbm_sc = lgb.LGBMRegressor(random_state=42).fit(X_train_fold, y_searches_pre_train.iloc[train_idx])
    train_processed.loc[train_processed.index[val_idx], 'searches_at_dbd15_pred'] = lgbm_sc.predict(X_val_fold)

# Train pre-models on the full training data to predict on the test set
print("  Training pre-models on full data for test set prediction...")
lgbm_seats_final = lgb.LGBMRegressor(random_state=42).fit(X_pre_train, y_seats_pre_train)
lgbm_searches_final = lgb.LGBMRegressor(random_state=42).fit(X_pre_train, y_searches_pre_train)
test_processed['seats_at_dbd15_pred'] = lgbm_seats_final.predict(X_test_pre_model).clip(min=0)
test_processed['searches_at_dbd15_pred'] = lgbm_searches_final.predict(X_test_pre_model).clip(min=0)

# Create the final meta-feature
train_processed['search_to_book_ratio_pred'] = train_processed['seats_at_dbd15_pred'] / (train_processed['searches_at_dbd15_pred'] + 1e-6)
test_processed['search_to_book_ratio_pred'] = test_processed['seats_at_dbd15_pred'] / (test_processed['searches_at_dbd15_pred'] + 1e-6)
print("Leakage-free meta-features generated for both train and test sets.")


# --- Step 5: Final ENSEMBLE Model Training & Submission ---
print("\n--- Step 5: Training final model and creating submission... ---")
final_model_features = pre_model_features + ['seats_at_dbd15_pred', 'searches_at_dbd15_pred', 'search_to_book_ratio_pred']
X_final_train = train_processed[final_model_features]
y_final_train = train_processed['final_seatcount']
X_final_test = test_processed[final_model_features]

# Parameters
lgbm_params = { 'objective': 'regression_l1', 'metric': 'rmse', 'random_state': 42, 'n_jobs': -1, 'n_estimators': 4810, 'learning_rate': 0.0279, 'num_leaves': 121, 'max_depth': 15, 'min_child_samples': 30, 'feature_fraction': 0.7166, 'bagging_fraction': 0.8765, 'bagging_freq': 1, 'lambda_l1': 3.46e-05, 'lambda_l2': 2.96e-06}
# ***FIX APPLIED HERE***: early_stopping_rounds is moved into the constructor params
xgb_params = { 'objective': 'reg:squarederror', 'eval_metric': 'rmse', 'random_state': 42, 'n_estimators': 2000, 'learning_rate': 0.02, 'max_depth': 8, 'subsample': 0.8, 'colsample_bytree': 0.8, 'n_jobs': -1, 'early_stopping_rounds': 50 }

# K-Fold Ensemble Training
kf = KFold(n_splits=5, shuffle=True, random_state=42)
lgbm_test_preds, xgb_test_preds = [], []

for fold, (train_idx, val_idx) in enumerate(kf.split(X_final_train)):
    print(f"--- Training Final Ensemble: FOLD {fold+1}/5 ---")
    X_train_fold, y_train_fold = X_final_train.iloc[train_idx], y_final_train.iloc[train_idx]
    X_val_fold, y_val_fold = X_final_train.iloc[val_idx], y_final_train.iloc[val_idx]

    print("  Training LGBM...")
    lgbm_model = lgb.LGBMRegressor(**lgbm_params)
    lgbm_model.fit(X_train_fold, y_train_fold, eval_set=[(X_val_fold, y_val_fold)], eval_metric='rmse', callbacks=[lgb.early_stopping(stopping_rounds=50, verbose=False)])
    lgbm_test_preds.append(lgbm_model.predict(X_final_test))

    print("  Training XGBoost...")
    xgb_model = xgb.XGBRegressor(**xgb_params)
    # ***FIX APPLIED HERE***: early_stopping_rounds is removed from .fit() call
    xgb_model.fit(X_train_fold, y_train_fold, eval_set=[(X_val_fold, y_val_fold)], verbose=False)
    xgb_test_preds.append(xgb_model.predict(X_final_test))

# --- Step 6: Blend Predictions and Submit ---
print("\n--- Step 6: Blending predictions and creating submission file... ---")
lgbm_final_preds = np.mean(lgbm_test_preds, axis=0)
xgb_final_preds = np.mean(xgb_test_preds, axis=0)
ensembled_predictions = 0.5 * lgbm_final_preds + 0.5 * xgb_final_preds
final_predictions = np.round(ensembled_predictions.clip(min=0)).astype(int)

submission_df = pd.DataFrame({'route_key': test_df['route_key'], 'final_seatcount': final_predictions})
submission_df.to_csv('submission_full_pipeline_ensemble.csv', index=False)

print("\n\n--- SCRIPT COMPLETE ---")
print("Submission file 'submission_full_pipeline_ensemble.csv' created successfully!")
print("\nSample of your submission:")
print(submission_df.head())
```
